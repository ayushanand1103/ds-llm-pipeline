```python
from datasets import load_dataset , concatenate_datasets
import os
```

```python
os.makedirs("data/pretrain", exist_ok=True)
```

```python
print("loading CodeParrot....")
code_parrot = load_dataset(
    "codeparrot/codeparrot-clean",
    split="train",
    streaming=True
)
```

```python
print("loading OpenWebMath....")
open_web_math = load_dataset(
    "open-web-math/open-web-math",
    split="train",
    streaming=True
)
```

```python
def save_samples(dataset, filename, n_samples=5000, text_field="content"):
    print(f"Saving {n_samples} samples to {filename}...")
    with open(filename, "w", encoding="utf-8") as f:
        count = 0
        for sample in dataset:
            if count >= n_samples:
                break
            text = sample.get(text_field, "").strip()
            if len(text) > 100:  # skip very short samples
                f.write(text + "\n\n")
                count += 1
    print(f"Done — saved {count} samples")
```

```python
save_samples(code_parrot,  "data/pretrain/codeparrot.txt",  n_samples=15000, text_field="content")
save_samples(open_web_math, "data/pretrain/openwebmath.txt", n_samples=10000, text_field="text")
```

```python
# ─── Combine only 2 datasets ────────────────────────────────────────────────────
print("Combining into corpus.txt...")
with open("data/pretrain/corpus.txt", "w", encoding="utf-8") as outfile:
    for fname in ["codeparrot.txt", "openwebmath.txt"]:
        path = f"data/pretrain/{fname}"
        if os.path.exists(path):
            with open(path, "r", encoding="utf-8") as infile:
                outfile.write(infile.read())

print("Done! corpus.txt ready.")
```
