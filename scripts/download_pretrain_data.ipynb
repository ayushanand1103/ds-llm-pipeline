```python
from datasets import load_dataset
import os
```

```python
base_path = "C:/datasets/llm-data/pretrain"
os.makedirs(base_path, exist_ok=True)

print("Data will be saved at:", base_path)
```

```python
print("Loading CodeParrot...")
code_parrot = load_dataset(
    "codeparrot/codeparrot-clean",
    split="train",
    streaming=True
)

print("Loading OpenWebMath...")
open_web_math = load_dataset(
    "open-web-math/open-web-math",
    split="train",
    streaming=True
)
```

```python
def save_samples(dataset, filename, n_samples=5000, text_field="content"):
    print(f"\nSaving {n_samples} samples to {filename}...")
    
    with open(filename, "w", encoding="utf-8") as f:
        count = 0
        
        for sample in dataset:
            if count >= n_samples:
                break

            text = sample.get(text_field, "").strip()
            text = text.replace("\n", " ")  # clean text

            if len(text) > 100:
                f.write(text + "\n\n")
                count += 1

                if count % 1000 == 0:
                    print(f"{count} samples saved...")

    print(f"Done — saved {count} samples")
```

```python
save_samples(
    code_parrot,
    f"{base_path}/codeparrot.txt",
    n_samples=15000,
    text_field="content"
)
```

```python
save_samples(
    open_web_math,
    f"{base_path}/openwebmath.txt",
    n_samples=10000,
    text_field="text"
)
```

```python
print("\nCombining into corpus.txt...")

with open(f"{base_path}/corpus.txt", "w", encoding="utf-8") as outfile:
    for fname in ["codeparrot.txt", "openwebmath.txt"]:
        path = f"{base_path}/{fname}"
        
        if os.path.exists(path):
            with open(path, "r", encoding="utf-8") as infile:
                for line in infile:
                    outfile.write(line)

print("Done! corpus.txt ready.")
```
