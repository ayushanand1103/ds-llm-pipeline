```python
from datasets import Dataset
import os
```

```python
CHUNKS_DIR = "C:/datasets/llm-data/pretrain/chunks"
OUTPUT_DIR = "C:/datasets/llm-data/pretrain/hf_dataset"
```

```python
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
print("Read all chunks...")
all_texts = []
chunk_files = sorted(os.listdir(CHUNKS_DIR))
```

```python
for i, fname in enumerate(chunk_files):
    if fname.endswith(".txt"):
        with open(os.path.join(CHUNKS_DIR, fname), "r", encoding="utf-8") as f:
            text = f.read().strip()
            if len(text) > 50:
                all_texts.append({"text": text})

    if i % 500 == 0:
        print(f"Read {i}/{len(chunk_files)} chunks...")

print(f"Total samples: {len(all_texts):,}")
```

```python
print(f"Total samples: {len(all_texts):,}")

# ─── Convert to HuggingFace Dataset ────────────────────────────────────────────
print("Converting to HuggingFace dataset...")
dataset = Dataset.from_list(all_texts)

# ─── Train / Val split ──────────────────────────────────────────────────────────
split = dataset.train_test_split(test_size=0.05, seed=42)
print(f"Train samples: {len(split['train']):,}")
print(f"Val samples:   {len(split['test']):,}")

# ─── Save to disk ───────────────────────────────────────────────────────────────
print("Saving to disk...")
split.save_to_disk(OUTPUT_DIR)
print(f"Done! HuggingFace dataset saved to {OUTPUT_DIR}")
```
