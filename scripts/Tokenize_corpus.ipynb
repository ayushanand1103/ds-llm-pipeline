```python
from transformers import AutoTokenizer
from datasets import load_from_disk
import os
```

```python
TOKENIZER_PATH = r"C:\datasets\llm-data\models\phi3-base"
DATASET_PATH   = r"C:\datasets\llm-data\pretrain\hf_dataset"
OUTPUT_PATH    = r"C:\datasets\llm-data\pretrain\tokenized_dataset"
MAX_LENGTH     = 512
```

```python
print("Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(
    TOKENIZER_PATH,
    local_files_only=True
)
```

```python
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
```

```python
print("Loading dataset...")
dataset = load_from_disk(DATASET_PATH)
print(dataset)
```

```python
def tokenize(batch):
    return tokenizer(
        batch["text"],
        truncation=True,
        max_length=MAX_LENGTH,
        padding="max_length",
        return_tensors=None
    )
```

```python
print("Tokenizing train split...")
tokenized_train = dataset["train"].map(
    tokenize,
    batched=True,
    batch_size=1000,
    remove_columns=["text"],
    desc="Tokenizing train"
)
```

```python
print("Tokenizing val split...")
tokenized_val = dataset["test"].map(
    tokenize,
    batched=True,
    batch_size=1000,
    remove_columns=["text"],
    desc="Tokenizing val"
)
```

```python
print("Saving...")
os.makedirs(OUTPUT_PATH, exist_ok=True)
tokenized_train.save_to_disk(os.path.join(OUTPUT_PATH, "train"))
tokenized_val.save_to_disk(os.path.join(OUTPUT_PATH, "val"))
```

```python
print(f"Done!")
print(f"Train: {len(tokenized_train):,} samples")
print(f"Val:   {len(tokenized_val):,} samples")
```
