```python
import os 
from transformers import AutoTokenizer , BitsAndBytesConfig , AutoModelForCausalLM
from peft import PeftModel
import torch , gc
os.environ["USE_TF"]="0"
os.environ["TRANSFORMERS_NO_TF"]="1"
```

```python
gc.collect()
torch.cuda.empty_cache()
```

```python
BASE_PATH = "C:\datasets\llm-data\models\phi3-pretrained-merged"
CHECKPOINT_PATH = "C:\datasets\llm-data\models\phi3-sft\checkpoint-87"
MERGED_PATH = "C:\datasets\llm-data\models\phi3-sft-phase1-merged"
```

```python
model = AutoModelForCausalLM.from_pretrained(BASE_PATH , torch_dtype = torch.float16 , device_map={"": 0}, local_files_only=True)
```

```python
model = PeftModel.from_pretrained(model , CHECKPOINT_PATH)
```

```python
model = model.merge_and_unload()
```

```python
tokenizer = AutoTokenizer.from_pretrained(BASE_PATH, local_files_only=True)
model.save_pretrained(MERGED_PATH)
tokenizer.save_pretrained(MERGED_PATH)
```
