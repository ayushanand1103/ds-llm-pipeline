```python
import os
```

```python
INPUT_FILE = "C:/datasets/llm-data/pretrain/corpus.txt"
OUTPUT_DIR = "C:/datasets/llm-data/pretrain/chunks"
CHUNK_SIZE = 1000
```

```python
os.makedirs(OUTPUT_DIR, exist_ok=True)

print("Reading corpus...")
with open(INPUT_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

print(f"Total lines: {len(lines):,}")
```

```python
chunks = [lines[i:i+CHUNK_SIZE] for i in range(0, len(lines), CHUNK_SIZE)]
print(f"Total chunks: {len(chunks):,}")
```

```python
for idx, chunk in enumerate(chunks):
    chunk_path = os.path.join(OUTPUT_DIR, f"chunk_{idx:04d}.txt")
    with open(chunk_path, "w", encoding="utf-8") as f:
        f.writelines(chunk)

    if idx % 500 == 0:
        print(f"Saved chunk {idx}/{len(chunks)}...")

print(f"Done! {len(chunks)} chunks saved to {OUTPUT_DIR}/")
```
