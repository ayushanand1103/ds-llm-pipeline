```python
import os 
os.environ["USE_TF"]="0"
os.environ["TRANSFORMERS_NO_TF"]="1" 

from transformers import AutoTokenizer , AutoModelForCausalLM , BitsAndBytesConfig
from peft import PeftModel
import torch , gc
```

```python
gc.collect()
torch.cuda.empty_cache()
```

```python
BASE_PATH       = r"C:\datasets\llm-data\models\phi3-base"
CHECKPOINT_PATH = r"C:\datasets\llm-data\models\phi3-pretrained\checkpoint-295"
MERGED_PATH     = r"C:\datasets\llm-data\models\phi3-pretrained-merged"
```

```python
bnb_config = BitsAndBytesConfig(load_in_8bit=True)
model = AutoModelForCausalLM.from_pretrained(BASE_PATH , quantization_config = bnb_config , device_map={"": 0}, local_files_only=True)
```

```python
model = PeftModel.from_pretrained(model , CHECKPOINT_PATH)
```

```python
model = model.merge_and_unload()
```

```python
tokenizer = AutoTokenizer.from_pretrained(BASE_PATH, local_files_only=True)
model.save_pretrained(MERGED_PATH)
tokenizer.save_pretrained(MERGED_PATH)
```
